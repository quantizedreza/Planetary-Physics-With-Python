
## Correlation between Salinity and Temperature


(C) Reza Rahemi - Scripps Institution Of Oceanography



```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

data = sio.loadmat('data.mat')

lat = data['argo_lat'].flatten()  # Latitude
lon = data['argo_lon'].flatten()  # Longitude 
pres = data['argo_pressure'].flatten()  # Pressure 
temp = data['argo_temperature']  # Temperature  2D array [lat, pres]
sal = data['argo_salinity']  # Salinity , 2D array [lat, pres]

temp_transect = temp  
sal_transect = sal    

Lat, Pres = np.meshgrid(lat, pres) #meshgrid for contour plotting

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#Plotting the temperature 

cf1 = ax1.contourf(Lat, Pres, temp_transect.T, cmap=cm.viridis, levels=20)
ax1.contour(Lat, Pres, temp_transect.T, colors='k', levels=10, linewidths=0.5)
ax1.set_ylim(pres.max(), pres.min())  
ax1.set_xlabel('Latitude (N)')
ax1.set_ylabel('Pressure (dbar)')
ax1.set_title('Temperature (C) at 340.5 E')
plt.colorbar(cf1, ax=ax1, label='Temperature (C)')

cf2 = ax2.contourf(Lat, Pres, sal_transect.T, cmap=cm.plasma, levels=20)
ax2.contour(Lat, Pres, sal_transect.T, colors='k', levels=10, linewidths=0.5)
ax2.set_ylim(pres.max(), pres.min())  
ax2.set_xlabel('Latitude (N)')
ax2.set_ylabel('Pressure (dbar)')
ax2.set_title('Salinity (psu) at 340.5 E')
plt.colorbar(cf2, ax=ax2, label='Salinity (psu)')

plt.tight_layout()
plt.savefig('transect_plot.png', dpi=300)
plt.savefig('transect_plot.pdf', dpi=300)
plt.show()
```
