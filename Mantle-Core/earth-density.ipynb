```python
# Earth Mass is M
#M = 5.972e24 # kg
# Earhh Radius is a and moment of inertia is I
#I/(a**2)=1.975e24 # kg

#M=4pi * integral of (r^2 * rho(r) dr) from 0 to a
# I/a^2 = 8pi/3 * integral of (r^4 * rho(r) dr) from 0 to a
#This gives rho(r) = 5.15e3 kg/m^3

#On the other hand, moment of intertia for uniform sphere gives I/a^2 = 8pi/15 * integral of (r^4 * rho dr) from 0 to a=0.4 M
# the solution to rho is 1.29e3 kg/m^3, which is much less than the previous value.

# This means that the Earth is not a uniform sphere, but has a denser core and less dense mantle and crust.Since
#the two integrals are different, the Earth must have a non-uniform density distribution. 

# So let's make a more complicated model. 

#M=4pi/3*c**3*rho_core + 4pi /3*(a**3-c**3)*rho_mantle 
# where c is the radius of the core, rho_core is the density of the core, and rho_mantle is the density of the mantle.

#Also, I= 8pi/15*c**5*rho_core + 8pi/15*(a**5-c**5)*rho_mantle


#These equations are in the form that the data vecotor is dT = [M, I/a^2], the model vector is mT = [rho_core, rho_mantle], and the kernel matrix is G = [[4pi/3*c**3, 4pi/3*(a**3-c**3)], [8pi/15*c**5, 8pi/15*(a**5-c**5)]]
# We can solve for the model vector mT by inverting the kernel matrix G and multiplying it by the data vector dT.
#the eauations of condition matrix is A = [[4pi/3*c**3, 4pi/3*(a**3-c**3)], [8pi/15*c**5, 8pi/15*(a**5-c**5)]]


#Note that A depends on ly on the two raddii , not on densitites nor the data. So we can solve for the model vector mT by inverting the matrix A and multiplying it by the data vector dT.  

#Here is a python code to solve for the model vector mT:






```

```python
#the rho_core is 12.5 g/cm^3 and the rho_mantle is 4.5 g/cm^3, which are consistent with the known values for the Earth's core and mantle densities.

#this is how to get them: 

import numpy as np
from scipy.linalg import inv
# Define the data vector
M = 5.972e24 # kg
I_over_a2 = 1.975e24 # kg
dT = np.array([M, I_over_a2])   
# Define the kernel matrix
a = 6371e3 # m, Earth's radius
c = 3480e3 # m, Earth's core radius
G = np.array([[4*np.pi/3*c**3, 4*np.pi/3*(a**3-c**3)], [8*np.pi/15*(c**5/a**2), 8*np.pi/15*(a**3-c**5/a**2)]])
# Invert the kernel matrix
G_inv = inv(G)
# Solve for the model vector
mT = G_inv @ dT
rho_core = mT[0] # kg/m^3
rho_mantle = mT[1] # kg/m^3
print("Density of the core:", rho_core, "kg/m^3")
print("Density of the mantle:", rho_mantle, "kg/m^3")
```
