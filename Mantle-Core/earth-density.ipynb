# Density Of The Earth - simple vs. realistic model

Earth mass:  
$M = 5.972 \times 10^{24}$ kg

Earth radius $a$, reduced moment of inertia:  
$\frac{I}{a^2} = 1.975 \times 10^{24}$ kg

Mass integral:  
$M = 4\pi \int_0^a r^2 \rho(r) \, dr$

Moment of inertia integral (scaled):  
$\frac{I}{a^2} = \frac{8\pi}{3} \int_0^a r^4 \rho(r) \, dr$

If we assume **constant density** $\rho(r) = \rho_0$:

- From mass → $\rho_0 = \dfrac{3M}{4\pi a^3} \approx 5.15 \times 10^3$ kg/m³  
- From moment of inertia → $\dfrac{I}{a^2} = \dfrac{2}{5}M \quad \Rightarrow \quad \rho_0 \approx 1.29 \times 10^3$ kg/m³ (much lower)

The strong discrepancy shows that **Earth cannot have uniform density**.  
It must have a **denser core** and **less dense mantle/crust**.

### Two-layer model (core + mantle)

Let  
- $c$ = core radius  
- $\rho_c$ = core density  
- $\rho_m$ = mantle density  

Then:

$$
M = \frac{4\pi}{3} c^3 \rho_c + \frac{4\pi}{3} (a^3 - c^3) \rho_m
$$

$$
I = \frac{8\pi}{15} c^5 \rho_c + \frac{8\pi}{15} (a^5 - c^5) \rho_m
$$

or equivalently (dividing the second equation by $a^2$):

$$
\frac{I}{a^2} = \frac{8\pi}{15} c^5 \rho_c + \frac{8\pi}{15} (a^5 - c^5) \rho_m \quad \Big/ \ a^2
$$

### Linear inverse problem formulation

Data vector:  
$\mathbf{d} = \begin{bmatrix} M \\ I/a^2 \end{bmatrix}$

Model vector:  
$\mathbf{m} = \begin{bmatrix} \rho_c \\ \rho_m \end{bmatrix}$

Kernel (design) matrix $\mathbf{G}$ (or $\mathbf{A}$):

$$
\mathbf{G} = \begin{bmatrix}
\frac{4\pi}{3} c^3 & \frac{4\pi}{3} (a^3 - c^3) \\[1.2ex]
\frac{8\pi}{15} \frac{c^5}{a^2} & \frac{8\pi}{15} \frac{a^5 - c^5}{a^2}
\end{bmatrix}
= \begin{bmatrix}
\frac{4\pi}{3} c^3 & \frac{4\pi}{3} (a^3 - c^3) \\[1.2ex]
\frac{8\pi}{15} c^5 / a^2 & \frac{8\pi}{15} (a^5 - c^5) / a^2
\end{bmatrix}
$$

Solution (when $\mathbf{G}$ is invertible):

$$
\mathbf{m} = \mathbf{G}^{-1} \mathbf{d}
$$

**Note**: The matrix $\mathbf{G}$ depends *only* on the geometrical parameters $a$ and $c$ — not on the densities or the data values.  
Once $c$ is assumed (or estimated), we can directly invert $\mathbf{G}$ and solve for $\rho_c$ and $\rho_m$.

```python
#the rho_core is 12.5 g/cm^3 and the rho_mantle is 4.5 g/cm^3, which are consistent with the known values for the Earth's core and mantle densities.

#this is how to get them: 

import numpy as np
from scipy.linalg import inv
# Define the data vector
M = 5.972e24 # kg
I_over_a2 = 1.975e24 # kg
dT = np.array([M, I_over_a2])   
# Define the kernel matrix
a = 6371e3 # m, Earth's radius
c = 3480e3 # m, Earth's core radius
G = np.array([[4*np.pi/3*c**3, 4*np.pi/3*(a**3-c**3)], [8*np.pi/15*(c**5/a**2), 8*np.pi/15*(a**3-c**5/a**2)]])
# Invert the kernel matrix
G_inv = inv(G)
# Solve for the model vector
mT = G_inv @ dT
rho_core = mT[0] # kg/m^3
rho_mantle = mT[1] # kg/m^3
print("Density of the core:", rho_core, "kg/m^3")
print("Density of the mantle:", rho_mantle, "kg/m^3")
```
